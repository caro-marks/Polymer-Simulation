```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
sns.set()
```

```python
beads_range=[x for x in range(10,50,4) if x not in (38,46)]
frac_range=[x/100 for x in range(0,100,25) if x!=0]
```

```python
columns=[]
with open('configs/configurations_of_10_beads.txt','r') as raw:
    cols=raw.readline()
    cols=cols.replace('|',',')
    cols=cols.replace(' ','')
    cols=cols.replace('\n','')
    for i in cols.split(','):
        columns.append(i)
```

```python
# data=pd.read_csv('configurations_of_10_beads.txt', sep='|');
# data.head(10)
```

```python
def GyraPerDens_Graphs(mons,fracs):
    for bead in mons:
        data=pd.read_csv('configs/configurations_of_'+str(bead)+'_beads.txt', sep='|')
        data.columns = columns
        dens=data['Density'].unique()
        gyra2=data.query('RealFracPhobic == "'+str(int(bead*(1-fracs[0])))+'"')['AverageRadiusofGyration']
        gyra5=data.query('RealFracPhobic == "'+str(int(bead*(1-fracs[1])))+'"')['AverageRadiusofGyration']
        gyra7=data.query('RealFracPhobic == "'+str(int(bead*(1-fracs[2])))+'"')['AverageRadiusofGyration']
        plt.plot(dens, gyra2, label=r'$f_{A}$ ~ 0.2')
        plt.plot(dens, gyra5, label=r'$f_{A}$ ~ 0.5')
        plt.plot(dens, gyra7, label=r'$f_{A}$ ~ 0.7')
        plt.xlabel("Density"); plt.ylabel("Average Radius of Gyration")
        plt.legend(loc=0, ncol=1); plt.title(str(bead)+"-beads-per-chain")
        plt.savefig("results/configuration/RadiusOfGyration/"+str(bead)+"-beads.png")
        plt.show()
```

```python
def DistPerDens_Graphs(mons):
    for bead in mons:
        data=pd.read_csv('configs/configurations_of_'+str(bead)+'_beads.txt', sep='|')
        data.columns = columns
        dens=data['Density'].unique()
        dist2=data.query('RealFracPhobic == "'+str(int(bead*(1-frac_range[0])))+'"')['MeanEnd-To-EndDistanceofChains']
        dist5=data.query('RealFracPhobic == "'+str(int(bead*(1-frac_range[1])))+'"')['MeanEnd-To-EndDistanceofChains']
        dist7=data.query('RealFracPhobic == "'+str(int(bead*(1-frac_range[2])))+'"')['MeanEnd-To-EndDistanceofChains']
        plt.plot(dens, dist2, label=r'$f_{A}$ ~ 0.2')
        plt.plot(dens, dist5, label=r'$f_{A}$ ~ 0.5')
        plt.plot(dens, dist7, label=r'$f_{A}$ ~ 0.7')
        plt.xlabel("Density"); plt.ylabel("Mean End-to-End Distance of Chains")
        plt.legend(loc=0, ncol=1); plt.title(str(bead)+"-beads-per-chain")
        plt.savefig("results/configuration/MeanEndToEndDistance/"+str(bead)+"-beads.png")
        plt.show()
```

```python
GyraPerDens_Graphs(beads_range,frac_range)
```

```python
DistPerDens_Graphs(beads_range)
```
