```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns; sns.set()
```

```python
Beads_range=[x for x in range(10,50,4) if x not in (38,46)] # # # #    monomer's range
Dens_range=[x/100 for x in range(0,81,5) if x>0] # # # #    dens' range
Frac_range=[int((1-f)*10%10) for f in [x/100 for x in range(0,100,25) if x!=0]] # # # #    frac' range
Frac_range.reverse()
pairs = {'0-0':'0.2', '0-1':'-0.4', '1-1':'-0.4'}
```

```python
def Data_2(bead, frac, dens):
    data = pd.read_table('./results/snapshots/coordination_analysis/f_'+str(frac)+'/b_'+str(bead)+'/d_'+str(dens)+'.txt', sep='\s+', header = 1)
    data.columns = ['Pair Separation Distance', '0-0', '0-1','1-1','NaN']
    data = data.drop(columns='NaN')
    return data
```

```python
def fftER(Beads, Fracs, Dens, Pairs):
    
    for pair in Pairs:
        
        for Bead in Beads:
            
            for Frac in Fracs:
                
                x = np.fft.fftfreq(Data_2(Bead, Frac, Dens[0]).shape[0])
                m = x > 0

#                 y0=1+Dens[0]*np.abs(np.fft.fft(Data_2(Bead, Frac, Dens[0])[pair])[m] -1)/Data_2(Bead, Frac, Dens[0]).shape[0]; plt.plot(x[m], y0, label = Dens[0]);
                y0=1+Dens[0]*np.abs(np.fft.fft(Data_2(Bead, Frac, Dens[0])[pair])[m]-1); plt.plot(x[m], y0, label = Dens[0]);
                y1=1+Dens[1]*np.abs(np.fft.fft(Data_2(Bead, Frac, Dens[1])[pair])[m]-1); plt.plot(x[m], y1, label = Dens[1]);
                y2=1+Dens[2]*np.abs(np.fft.fft(Data_2(Bead, Frac, Dens[2])[pair])[m]-1); plt.plot(x[m], y2, label = Dens[2]);
                y3=1+Dens[3]*np.abs(np.fft.fft(Data_2(Bead, Frac, Dens[3])[pair])[m]-1); plt.plot(x[m], y3, label = Dens[3]);
                y4=1+Dens[4]*np.abs(np.fft.fft(Data_2(Bead, Frac, Dens[4])[pair])[m]-1); plt.plot(x[m], y4, label = Dens[4]);
                y5=1+Dens[5]*np.abs(np.fft.fft(Data_2(Bead, Frac, Dens[5])[pair])[m]-1); plt.plot(x[m], y5, label = Dens[5]);
                y6=1+Dens[6]*np.abs(np.fft.fft(Data_2(Bead, Frac, Dens[6])[pair])[m]-1); plt.plot(x[m], y6, label = Dens[6]);
                y7=1+Dens[7]*np.abs(np.fft.fft(Data_2(Bead, Frac, Dens[7])[pair])[m]-1); plt.plot(x[m], y7, label = Dens[7]);
                y8=1+Dens[8]*np.abs(np.fft.fft(Data_2(Bead, Frac, Dens[8])[pair])[m]-1); plt.plot(x[m], y8, label = Dens[8]);
                y9=1+Dens[9]*np.abs(np.fft.fft(Data_2(Bead, Frac, Dens[9])[pair])[m]-1); plt.plot(x[m], y9, label = Dens[9]);
                y10=1+Dens[10]*np.abs(np.fft.fft(Data_2(Bead, Frac, Dens[10])[pair])[m]-1); plt.plot(x[m], y10, label = Dens[10]);
                y11=1+Dens[11]*np.abs(np.fft.fft(Data_2(Bead, Frac, Dens[11])[pair])[m]-1); plt.plot(x[m], y11, label = Dens[11]);
                y12=1+Dens[12]*np.abs(np.fft.fft(Data_2(Bead, Frac, Dens[12])[pair])[m]-1); plt.plot(x[m], y12, label = Dens[12]);
                y13=1+Dens[13]*np.abs(np.fft.fft(Data_2(Bead, Frac, Dens[13])[pair])[m]-1); plt.plot(x[m], y13, label = Dens[13]);
                y14=1+Dens[14]*np.abs(np.fft.fft(Data_2(Bead, Frac, Dens[14])[pair])[m]-1); plt.plot(x[m], y14, label = Dens[14]);
                y15=1+Dens[15]*np.abs(np.fft.fft(Data_2(Bead, Frac, Dens[15])[pair])[m]-1); plt.plot(x[m], y15, label = Dens[15]);                 
                
                plt.xlabel("q"); plt.ylabel('S(q) '+str(pair));
                plt.title('static structure aproximation (beads: '+str(Bead)+"; frac: "+str(Frac)+")")
                plt.legend(loc=0, ncol=2);
                plt.savefig('./results/snapshots/coordination_analysis/f_'+str(Frac)+'/b_'+str(Bead)+'/p_'+str(pair)+'-(fft).png');
                plt.show()
```

```python
def fftER_2(Beads, Fracs, Denss, Pairs):
    
    for pair in Pairs:
        
        for Frac in Fracs:
            
            for Dens in Denss:
                
                x = np.fft.fftfreq(Data_2(Beads[0], Frac, Dens).shape[0])
                m = x > 0

                y0=1+Dens*np.abs(np.fft.fft(Data_2(Beads[0], Frac, Dens)[pair])[m]-1); plt.plot(x[m], y0, label = Beads[0]);
                y1=1+Dens*np.abs(np.fft.fft(Data_2(Beads[1], Frac, Dens)[pair])[m]-1); plt.plot(x[m], y0, label = Beads[1]);
                y2=1+Dens*np.abs(np.fft.fft(Data_2(Beads[2], Frac, Dens)[pair])[m]-1); plt.plot(x[m], y0, label = Beads[2]);
                y3=1+Dens*np.abs(np.fft.fft(Data_2(Beads[3], Frac, Dens)[pair])[m]-1); plt.plot(x[m], y0, label = Beads[3]);
                y4=1+Dens*np.abs(np.fft.fft(Data_2(Beads[4], Frac, Dens)[pair])[m]-1); plt.plot(x[m], y0, label = Beads[4]);
                y5=1+Dens*np.abs(np.fft.fft(Data_2(Beads[5], Frac, Dens)[pair])[m]-1); plt.plot(x[m], y0, label = Beads[5]);
                y6=1+Dens*np.abs(np.fft.fft(Data_2(Beads[6], Frac, Dens)[pair])[m]-1); plt.plot(x[m], y0, label = Beads[6]);
                y7=1+Dens*np.abs(np.fft.fft(Data_2(Beads[7], Frac, Dens)[pair])[m]-1); plt.plot(x[m], y0, label = Beads[7]);
                
                plt.xlabel("q"); plt.ylabel('S(q) '+str(pair));
                plt.title('static structure aproximation (Dens: '+str(Dens)+"; frac: "+str(Frac)+")")
                plt.legend(loc=0, ncol=2);
                plt.savefig('./results/snapshots/coordination_analysis/f_'+str(Frac)+'/d_'+str(Dens)+'/p_'+str(pair)+'-(fft).png');
                plt.show()
```
