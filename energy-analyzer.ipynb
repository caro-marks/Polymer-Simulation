```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
sns.set()
```

```python
beads_range=[x for x in range(10,50,4) if x not in (38,46)] # # # #    monomer's range
dens_range=[x/100 for x in range(0,81,5) if x>0] # # # #    dens' range
frac_range=[x/100 for x in range(0,100,25) if x!=0] # # # #    frac' range
energies=['Total','Kinetic','Bonded','Non-Bonded','Pressure']
```

```python
def Data(frac,bead,dens):
    data = pd.read_csv ('./frac_'+str(frac)+'/m_'+str(bead)+'/d_'+str(dens)+'/energy.csv')
    return data
```

```python
def listCreater(fracs, beads, denss, Energy):
    ls = [];
    for frac in fracs:
        l = [];
        for dens in denss:
            try:
                l.append(Data(frac,beads,dens)[str(Energy)].mean())
            except:
                continue
        ls.append(l)
    return ls
```

```python
def energyGraph(Mons,Fracs,Dens,Energy):
    for Beads in Mons:
        x = Dens;
        y1=listCreater(Fracs, Beads, Dens, Energy)[2];
        y2=listCreater(Fracs, Beads, Dens, Energy)[1];
        y3=listCreater(Fracs, Beads, Dens, Energy)[0];
        plt.plot(x, y1, label=r'$f_{A}$ ~ 0.2');
        plt.plot(x, y2, label=r'$f_{A}$ ~ 0.5');
        plt.plot(x, y3, label=r'$f_{A}$ ~ 0.7');
        plt.xlabel("Density"); plt.ylabel(str(Energy));
        plt.title(str(Beads)+"-beads-per-chain");
        plt.legend(loc=0, ncol=1);
        plt.savefig("results/energies/"+str(Energy)+"/"+str(Beads)+"-beads.png");
        plt.show()
```

```python
for energy in energies:
    energyGraph(beads_range,frac_range,dens_range,energy)
```
