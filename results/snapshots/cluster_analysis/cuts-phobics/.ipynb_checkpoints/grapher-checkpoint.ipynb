```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
cut_range = tuple([(i+10)/10 for i in range(1,11)])
dens_range = tuple([i/100 for i in range(5,85,5)])
beads_range = tuple([i for i in range(10,50, 4) if i not in (38,46)])
```

```python
for i in cut_range:
    data = pd.read_csv(f'{i}/Frac_0.25')
    plt.plot(dens_range, data['10'],  label ='10')
    plt.plot(dens_range, data['14'],  label ='14')
    plt.plot(dens_range, data['18'],  label ='18')
    plt.plot(dens_range, data['22'],  label ='22')
    plt.plot(dens_range, data['26'],  label ='26')
    plt.plot(dens_range, data['30'],  label ='30')
    plt.plot(dens_range, data['34'],  label ='34')
    plt.plot(dens_range, data['42'],  label ='42')
    plt.xlabel('densities'); plt.ylabel('clusters counts')
    plt.title(f'cut_range = {i}')
    plt.savefig(f'plots/25/{i}.png')
    plt.show()
```
